This project involved analyzing mosquito attraction data using statistical analysis and data visualization techniques in Python, concluding that beer consumption appears to correlate with higher mosquito attraction based on descriptive statistics and hypothesis testing via simulation.

Skills: Bootstrap Sampling · Python (Programming Language) · Statistical Data Analysis · Data Visualization


```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the CSV file
df = pd.read_csv('mosquitos_data.csv')

# Display the DataFrame
print(df)
```

```python
# Set the style of seaborn
sns.set(style="whitegrid")

# Create the boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x="Treatment", y="Response", data=df, palette="Set2")

# Add title and labels
plt.title("Number of Mosquitos in Each Group (Beer vs Water)")
plt.xlabel("Treatment")
plt.ylabel("Number of Mosquitos")

# Show the plot
plt.show()
```

Based on the graph, there is a clear indication that mosquitoes were more attracted to beer drinkers compared to non-beer drinkers. The data reveals that the number of mosquito landings or bites was significantly higher for individuals who consumed beer. This suggests a strong association between beer consumption and increased attractiveness to mosquitoes.

```python
summary_stats = df.groupby('Treatment')['Response'].describe()

print(summary_stats)
```

```python
# Creating a DataFrame from the given data
data = {
    'Response': [27, 20, 21, 26, 27, 31, 24, 21, 20, 19, 23, 24, 28, 19, 24, 29, 18, 20, 17, 31, 20, 25, 28, 21, 27, 21, 22, 15, 12, 21, 16, 19, 15, 22, 24, 19, 23, 13, 22, 20, 24, 18, 20],
    'Treatment': ['Beer']*25 + ['Water']*18
}

df = pd.DataFrame(data)

# Extracting 'Response' column as an array
mosquito_counts = df['Response'].values

# Number of iterations
N = 10000

# Initial observed difference
observed_diff = 4.4

# List to store results of experiments
results = []

# Counting number of times the experiment outcome is 4.4 or greater
k = 0

# Simulation
for _ in range(N):
    # Shuffle the array items
    np.random.shuffle(mosquito_counts)

    # Split into two subgroups
    beer_sample = mosquito_counts[:25]
    water_sample = mosquito_counts[25:]

    # Compute the difference of means
    mean_diff = np.mean(beer_sample) - np.mean(water_sample)

    # Append the current mean to the results
    results.append(mean_diff)

    # Count number of times experiment outcome 4.4 or greater
    if mean_diff >= observed_diff:
        k += 1

# Calculate the probability
probability = k / N



# Plot histogram of the mean differences
plt.hist(results, bins=30, edgecolor='k', alpha=0.7)
plt.axvline(observed_diff, color='red', linestyle='dashed', linewidth=1)
plt.title('Histogram of Mean Differences from Simulated Experiments')
plt.xlabel('Mean Difference')
plt.ylabel('Frequency')
plt.show()
```

```python
# Print the results
print(f"Probability of obtaining the experiment outcome of 4.4 or greater: {probability}")
```
